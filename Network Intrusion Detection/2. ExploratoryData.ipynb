- Thực hiện mã hóa

```python
from sklearn.preprocessing import LabelEncoder

# Ham thực hiện mã hóa 
def le(df):
    for col in df.columns:
        if df[col].dtype == 'object':
                label_encoder = LabelEncoder()
                df[col] = label_encoder.fit_transform(df[col])


le(train)
le(test)
```

### Lọc ra các đặc trưng quan trọng

- Sử dụng phương pháp RFE (Recursive Feature Elimination) kết hợp với mô hình Random Forest để lựa chọn các đặc trưng quan trọng nhất từ dữ liệu huấn luyện. Mục tiêu của phương pháp này là xác định những đặc trưng có đóng góp lớn nhất vào việc dự đoán nhãn đầu ra (Y_train).

```python
from sklearn.ensemble import RandomForestClassifier  #RandomForestClassifier để sử dụng mô hình rừng ngẫu nhiên
from sklearn.feature_selection import RFE  #Recursive Feature Elimination (RFE) để lựa chọn đặc trưng
import itertools   #itertools để kết hợp các danh sách

# Khởi tạo mô hình Random Forest
rfc = RandomForestClassifier()

# Tạo bộ lọc RFE, chọn 10 đặc trưng quan trọng nhất
rfe = RFE(rfc, n_features_to_select=10)

# Huấn luyện RFE trên dữ liệu huấn luyện (X_train là các đặc trưng, Y_train là nhãn)
rfe = rfe.fit(X_train, Y_train)

# Tạo ánh xạ giữa các đặc trưng được chọn (True/False) và tên cột trong X_train
feature_map = [(i, v) for i, v in itertools.zip_longest(rfe.get_support(), X_train.columns)]

# Lọc ra danh sách các đặc trưng được chọn (những cột có giá trị True)
selected_features = [v for i, v in feature_map if i==True]

selected_features
```

- Như vậy ta có 10 đặc trưng được lựa chọn.
1. **`protocol_type`**:
   - Loại giao thức mạng được sử dụng trong kết nối, ví dụ như TCP, UDP hoặc ICMP. Đây là một trong những đặc trưng quan trọng trong việc phân loại loại lưu lượng mạng.

2. **`flag`**:
   - Cờ (flag) liên quan đến trạng thái của phiên giao dịch trong giao thức TCP. Các cờ như `SF`, `S0`, `REJ` biểu thị các trạng thái khác nhau của kết nối mạng.

3. **`src_bytes`**:
   - Số byte dữ liệu gửi từ máy nguồn (source) đến máy đích (destination). Cột này cho thấy lượng dữ liệu truyền tải từ phía nguồn, giúp phân biệt giữa các cuộc tấn công và lưu lượng mạng thông thường.

4. **`dst_bytes`**:
   - Số byte dữ liệu gửi từ máy đích (destination) đến máy nguồn (source). Tương tự như `src_bytes`, nhưng hướng ngược lại. Cột này thể hiện lượng dữ liệu trả lại từ máy đích đến máy nguồn.

5. **`count`**:
   - Tổng số kết nối đến máy đích từ cùng một nguồn trong một khoảng thời gian. Đây là một chỉ số quan trọng trong việc phát hiện các cuộc tấn công như **DoS** (Denial of Service), vì các cuộc tấn công này thường tạo ra nhiều kết nối từ cùng một nguồn trong một thời gian ngắn.

6. **`same_srv_rate`**:
   - Tỷ lệ các kết nối trong đó giao thức và cổng nguồn giống nhau với các kết nối đã thực hiện trước đó. Cột này phản ánh mức độ giống nhau giữa các kết nối trong cùng một dịch vụ. Tỷ lệ cao có thể chỉ ra lưu lượng mạng hợp pháp, trong khi tỷ lệ thấp có thể là dấu hiệu của một cuộc tấn công.

7. **`diff_srv_rate`**:
   - Tỷ lệ kết nối giữa các dịch vụ khác nhau. Cột này đo lường mức độ đa dạng của các kết nối đến các dịch vụ khác nhau, giúp phân biệt giữa lưu lượng mạng hợp pháp và các cuộc tấn công phức tạp.

8. **`dst_host_srv_count`**:
   - Số lượng kết nối đến các dịch vụ khác nhau của một máy đích. Thể hiện số lượng dịch vụ mà một máy đích đang phục vụ. Một số lượng kết nối lớn có thể là dấu hiệu của các cuộc tấn công hoặc hoạt động mạng bình thường.

9. **`dst_host_same_srv_rate`**:
   - Tỷ lệ kết nối đến các dịch vụ giống nhau trên máy đích. Giống như `same_srv_rate`, nhưng áp dụng cho máy đích. Tỷ lệ cao có thể chỉ ra các cuộc giao tiếp hợp pháp, trong khi tỷ lệ thấp có thể chỉ ra các hành vi đáng ngờ.

10. **`dst_host_same_src_port_rate`**:
    - Tỷ lệ các kết nối đến máy đích từ cùng một cổng nguồn. Cột này giúp theo dõi xem các kết nối từ cùng một nguồn có sử dụng cổng nguồn giống nhau hay không. Tỷ lệ cao có thể chỉ ra hành vi mạng bình thường hoặc tấn công.

### Chuẩn hóa và chia dữ liệu thành hai tập

```python
# Chuẩn hóa dữ liệu
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
X_train = scale.fit_transform(X_train)
test = scale.fit_transform(test)
```

```python
# Chia dữ liệu thành 2 tâp
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, train_size=0.70, random_state=2)
```
